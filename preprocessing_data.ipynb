For statistic

```python
df = pd.read_csv('output_data/global.csv')
```

```python
df.head()
```

```python
df = df.iloc[:, [0, 1, 2, 3]]
```

```python
df.head()
```

```python
df['active'] = df.cases - df.deaths - df.recovered
```

```python
df.head()
```

```python
df['death_rate'] = round(df.deaths/df.cases*100, 2)
```

```python
df.head()
```

```python
df['recovered_rate'] = round(df.recovered/df.cases*100, 2)
```

```python
df.head()
```

```python
df['flag'] = 'assets/flags/' + df.country + '.png'
```

```python
df.head()
```

```python
df['rank'] = df.index + 1
```

```python
df.head()
```

```python
df.to_csv('table.csv', index=False)
```
